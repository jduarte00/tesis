Creación de gráfica de series de tiempo para comparar distancia euclidana vs distancia basada en correlación

```python
pert_normal = np.random.rand(8)/8
```

```python
data = np.array([0.13, 0.22, 0.15, 0.05, 0.17, 0.33, 0.42, 0.13])
data2  = data + 0.30
data_pert = data + pert_normal
```

```python
plt.plot(data, label='blogs_read', 
         linewidth=1, linestyle = '--')
plt.plot(data2, label='blogs_read', 
         linewidth=1, linestyle = '--')
plt.plot(data_pert, label='blogs_read', 
         linewidth=1, linestyle = '--')
```
