```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
samples_50 = np.random.randn(50, 50)
samples_500 = np.random.randn(100, 50)
samples_5000 = np.random.randn(200, 50)
```

```python
corr_mat_500 = np.corrcoef(samples_50.T)
corr_mat_5000 = np.corrcoef(samples_500.T)
corr_mat_50000 = np.corrcoef(samples_5000.T)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 15))
sns.heatmap(corr_mat_500, cmap='coolwarm', vmin=-1, vmax=1, ax=axes[0])
axes[0].set_title(r'q = 1')
sns.heatmap(corr_mat_50000, cmap='coolwarm', vmin=-1, vmax=1, ax=axes[1])
axes[1].set_title(r'q = $\frac{1}{4}$')
```
